```python
import random

category_descriptions = {
    'Sandwich': ["delicious", "fresh", "tasty", "grilled", "satisfying"],
    'Burger': ["juicy", "flame-grilled", "hearty", "cheesy", "mouthwatering"],
    'Pizza': ["crispy", "topped", "cheesy", "savory", "oven-baked"],
    'Taco': ["zesty", "spicy", "crunchy", "authentic", "flavorful"],
    'Tenders': ["crispy", "tender", "finger-licking", "golden", "satisfying"],
    'Nuggets': ["crispy", "golden", "tasty", "dippable", "irresistible"],
    'Wrap': ["fresh", "light", "packed", "portable", "filling"],
    'Steak': ["juicy", "tender", "grilled", "sizzling", "mouthwatering"],
    'Hot Dog': ["classic", "grilled", "satisfying", "relished", "flavorful"],
    'Other': ["unique", "surprising", "innovative", "exciting", "delightful"]
}

def generate_category_description(category):
    return f"{random.choice(category_descriptions[category])} {category.lower()}"

for index, row in Menu.iterrows():
    item_name = row['ItemName'].lower()
    if 'sandwich' in item_name:
        Menu.at[index, 'Description'] = generate_category_description('Sandwich')
    elif 'burger' in item_name:
        Menu.at[index, 'Description'] = generate_category_description('Burger')
    elif 'pizza' in item_name:
        Menu.at[index, 'Description'] = generate_category_description('Pizza')
    elif 'taco' in item_name:
        Menu.at[index, 'Description'] = generate_category_description('Taco')
    elif 'tenders' in item_name:
        Menu.at[index, 'Description'] = generate_category_description('Tenders')
    elif 'nuggets' in item_name:
        Menu.at[index, 'Description'] = generate_category_description('Nuggets')
    elif 'wrap' in item_name:
        Menu.at[index, 'Description'] = generate_category_description('Wrap')
    elif 'steak' in item_name:
        Menu.at[index, 'Description'] = generate_category_description('Steak')
    elif 'hot dog' in item_name:
        Menu.at[index, 'Description'] = generate_category_description('Hot Dog')
    else:
        Menu.at[index, 'Description'] = generate_category_description('Other')
```

```python
combined_df['tags']=combined_df['category']+ ' '+combined_df['Description_x']+ ' '+combined_df['Description_y']+ ' '+combined_df['RestaurantLocation']+ ' '+combined_df['OpeningHours'] 
```

```python
data = combined_df[['Menuid','ItemName','tags']]
```

```python
data.count()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))
    return ' '.join(l)
```

```python
data['tags'] = data['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(data['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(name):
    index = data[data['ItemName'] == name].index[0]
    distances = similarity[index]
    
    
    listOfRecomendation = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:20]
    for i in listOfRecomendation:
        print(data.iloc[i[0]].ItemName)
recommend('Single Bacon Smokehouse Burger')
```
